````python
#Se importan las librerias necesarias
import yfinance as yf
import scipy
import pandas as pd
import numpy as np
from pysentimiento import create_analyzer
import warnings
warnings.filterwarnings("ignore")
import matplotlib.pyplot as plt
# libreria para interactuar con el interprete y el sistema operativo
import sys
import seaborn as sns
````

Se realiza la sacada de los datos financieros con la libreria yfinance correspondiente, para esto solo se van a sacar los datos de Isrrael y USA, porque los datos de Rusia estan incompletos en este provedor entonces utilizamos otro despues.


````python
#Se definen los indices de los parametros que se quieren llamar a la API de y finance
indices = {
    "USA_SP500": "^GSPC",
    "Israel_TA35": "TA35.TA"#,
    #"Russia_MOEX": "IMOEX.ME" - La parte de Rusia esta incompleta entonces se saca de otra fuente de datos
}
# Se define el periodo donde se quieren sacar los datos
start_date = "2023-01-03"
end_date = "2026-02-09"

#Se define una lista vacia donde se van a guardar los datos de los data frame
data = {}
# Se realiza un loop para pedir a la api de yfinance los datos necesarios y se guardan los df en la lista data
for country, ticker in indices.items():
    df = yf.download(ticker, start=start_date, end=end_date)
    df = df.reset_index()
    data[country] = df

# Se printea para conocer si se importo bien
for country, df in data.items():
    print(f"\n{country}")
    print(df.head())
````

Se realiza la sacada de datos de Rusia, esto se hizo manual en la pagina https://tradingeconomics.com/rusia/mercado-de-valores , a continuacion un codigo de la data que nos salio guardada en el repositorio

````python
url = "https://raw.githubusercontent.com/JRcepeda/Datos-Maestria/refs/heads/main/MOEX2.csv"
df_Russia = pd.read_csv(url,sep=";")

df_Russia.head()
````

Se realiza una union de toda la data en un df original, en este caso como solo se utiliza la data de Close entonces solo se guarda esta

````python
# Se crea un df apra cada país por facilidad donde solo se guarda la data y el parametro CLOSE
df_USA_ = data["USA_SP500"][["Date", "Close"]]
df_Israel_ = data["Israel_TA35"][["Date", "Close"]]

#Se transforma data en un type date
df_USA_['Date'] = pd.to_datetime(df_USA_['Date'])
df_Israel_['Date'] = pd.to_datetime(df_Israel_['Date'])

# Se establece una columna Date para guardar las fechas
df_USA_ = df_USA_.set_index('Date')
df_Israel_ = df_Israel_.set_index('Date')

# Se renombra la columna Close a el pais para guardarlo despues en el df final
df_USA_ = df_USA_.rename(columns={'Close': 'USA'})
df_Israel_ = df_Israel_.rename(columns={'Close': 'Israel'})


# Se prepara el DataFrame para Rusia
df_Russia_ = df_Russia.copy()

# Se seleccionan las columnas correctas: 'TRADEDATE' para la fecha y 'CLOSE' para el valor
df_Russia_ = df_Russia_[['TRADEDATE', 'CLOSE']].copy()

#Se renombran los indices anteriores con date y russia
df_Russia_.columns = ['Date', 'Russia']

# Convertir la columna 'Russia' (que contiene los valores 'CLOSE') a numérico, reemplazando la coma por un punto
df_Russia_['Russia'] = df_Russia_['Russia'].str.replace(',', '.').astype(float)

#Se convierte la columna date en el formato datetime de pandas
df_Russia_['Date'] = pd.to_datetime(
    df_Russia_['Date'],
    format='%d.%m.%Y'
)
# Se establece como indice principal del df como la columna date
df_Russia_ = df_Russia_.set_index('Date')

# Se concatenan los tres DataFrames  por columnas (axis=1) manteniendo las fechas que existen en los tres DataFrames
df_final = pd.concat(
    [df_USA_, df_Israel_, df_Russia_],
    axis=1,
    join='inner'   # solo fechas comunes
)
# Se ordena el data frame por fecha
df_final = df_final.sort_index()
#Se reinicia el incdice apra convertir date en columna
df_final = df_final.reset_index()
# Se renombra el indice a Date
df_final = df_final.rename(columns={'index': 'Date'})
````

### **2.2. Análisis estadístico descriptivo**

#### **Análisis cruzado (datos financieros + noticias)**

````python
#objeto para categorizar las noticias
analyzer = create_analyzer(task="sentiment", lang="es")
````

````python
#Se transforman los datos de noticias en términos numéricos
def label_news(row):
  #Los valores pueden ser ´POS´ que es sentimiento positivo, ´NEG´ que es sentimiento negativo y ´NEU´ que es sentimiento neutro
  resultados = {}
  for col in row.index:
      resultados[col] = analyzer.predict(row[col]).output
  return pd.Series(resultados)
````

````python
#Se transforma la columna 'Mes / Año' en tipo fecha para facilitar el análisis
df_news['Mes / Año'] = pd.to_datetime(df_news['Mes / Año'],dayfirst=True)
df_news['Mes / Año'] = df_news['Mes / Año']+pd.offsets.MonthEnd(0)
````

````python
df_news_desc = df_news.copy()
df_news_desc.set_index('Mes / Año',inplace=True)
df_sent =  df_news_desc.apply(label_news,axis=1)
````

````python
#gráfica análisis del sentimiento de las noticias encontradas
countries = df_sent.columns
fig,ax = plt.subplots(1,3,figsize=(30,15))
labels = ['POS','NEG','NEU']
colors = {
    'POS':'forestgreen',
    'NEG':'crimson',
    'NEU':'silver'
}
for j,i in enumerate(countries):
  df_sent[i].value_counts(normalize=True).plot.pie(autopct='%1.2f%%', figsize=(15, 10),ax=ax[j],colors = [colors[i] for i in df_sent[i].value_counts(normalize=True).index])
  ax[j].set_title(f"Sentimiento de noticias en: {i}")
  ax[j].axis('off')
plt.show()
````

Realizando un analisis de participaciones se observa que:
* En Rusia la mayoría de las noticias pueden ser concideradas neutras por el modelo de lenguaje, solo el 36% representan eventos políticos negativos.
* En Israel el 72.73% de la noticias se consideran negativas, esto se puede atribuir al la guerra que con palestina que enfrenta el país.
* En Estados Unidos la mayor parte de las noticias son neutras con una participación del 45.45%. El dato que difiere con las noticias de los otros paises es que la victoria de Trump en las elecciones con promesas de negociaciones de paz apunta hacia un sentimiento positivo en el panorama político del país, a esto se atribuye el 9.09% en participación de noticias.

````python
#se usa una copia y se establece la fecha como índice
df_final_desc = df_final.copy()
df_final_desc.set_index('Date',inplace=True)

# se calcula los dias bursatiles de cada mes
dias_mes = df_final_desc.resample('M').count()

#se calculan los rendimientos esperados
rendimientos_diarios = df_final_desc.pct_change().resample('M').mean()
rendimientos_esperados_mes = ((1+rendimientos_diarios)**dias_mes)-1

#se calcula la volatildad esperada
volatilidad_diaria = df_final_desc.pct_change().resample('M').std()
volatilidad_esperada_mes = volatilidad_diaria*(dias_mes**(1/2))

#reasignación de nombres de columnas
rendimientos_esperados_mes.columns =['Estados Unidos 🇺🇸','Israel 🇮🇱','Rusia 🇷🇺']
volatilidad_esperada_mes.columns =['Estados Unidos 🇺🇸','Israel 🇮🇱','Rusia 🇷🇺']
````

````python
#fusión de datasets
df_finance = volatilidad_esperada_mes.merge(rendimientos_esperados_mes,on='Date',suffixes=('_volatilidad', '_rendimiento'))
````

````python
#unión de rentabilidades, volativlidades y sentimiento
df_finance_and_sent = df_finance.merge(df_sent,left_on='Date',right_on='Mes / Año')
print(df_finance_and_sent.shape)
df_finance_and_sent.head()
````

````python
fig,ax = plt.subplots(1,3,figsize=(15,6))
for j,i in enumerate(['Estados Unidos 🇺🇸','Israel 🇮🇱','Rusia 🇷🇺']):
  df_finance_and_sent.groupby(i).agg({f'{i}_volatilidad':'mean',f'{i}_rendimiento':'mean'}).plot.bar(ax=ax[j])
  ax[j].set_title(i, fontsize=13, fontweight='bold')
  ax[j].set_xlabel("Sentimiento")
  ax[j].set_ylabel("Promedio")
  ax[j].grid(axis='y', linestyle='--', alpha=0.4)
plt.show()

````

````python
# CÁLCULO DE RENTABILIDAD


print("RENTABILIDAD ESPERADA MENSUAL")

df_final.columns = ['Date', 'USA', 'Israel', 'Russia']

# Retornos diarios,  pct_change() calcula: (precio_hoy - precio_ayer) / precio_ayer.

df_final['USA_daily_return']    = df_final['USA'].pct_change()
df_final['Israel_daily_return'] = df_final['Israel'].pct_change()
df_final['Russia_daily_return'] = df_final['Russia'].pct_change()


# Columna de agrupación mensual

df_final['Year_Month'] = df_final['Date'].dt.to_period('M')

# Rentabilidad esperada mensual, promedio de retornos diarios agrupados por mes
monthly_returns = df_final.groupby('Year_Month').agg({
    'USA_daily_return': 'mean',
    'Israel_daily_return': 'mean',
    'Russia_daily_return': 'mean'
}).reset_index()

monthly_returns.columns = ['Month', 'USA_monthly_exp', 'Israel_monthly_exp', 'Russia_monthly_exp']

#Conteo de días hábiles por mes

fechas = df_final.groupby('Year_Month').agg(
    USA_monthly_exp    = ('USA',    'count'),
    Israel_monthly_exp = ('Israel', 'count'),
    Russia_monthly_exp = ('Russia', 'count'),
).reset_index()

fechas.columns = ['Month', 'USA_monthly_exp', 'Israel_monthly_exp', 'Russia_monthly_exp']

## Filtrar solo columnas de valores
fechas_ = fechas[['USA_monthly_exp', 'Israel_monthly_exp', 'Russia_monthly_exp']]
monthly_returns_ = monthly_returns[['USA_monthly_exp', 'Israel_monthly_exp', 'Russia_monthly_exp']]

# Rendimiento esperado mensual, fórmula: (1 + r_diario)^n_dias - 1

rendimientos_esperados_mes = (1 + monthly_returns_) ** fechas_ - 1


print(rendimientos_esperados_mes)
````

````python
# ESTADÍSTICAS DESCRIPTIVAS - RENTABILIDAD ESPERADA MENSUAL


print("ESTADÍSTICAS DESCRIPTIVAS, RENTABILIDAD ESPERADA MENSUAL")

stats_rentabilidad = pd.DataFrame({
    'País': ['USA (S&P 500)', 'Israel (TA-35)', 'Rusia (MOEX)'],
    'Media (%)': [
        rendimientos_esperados_mes['USA_monthly_exp'].mean(),
        rendimientos_esperados_mes['Israel_monthly_exp'].mean(),
        rendimientos_esperados_mes['Russia_monthly_exp'].mean()
    ],
    'Mediana ': [
        rendimientos_esperados_mes['USA_monthly_exp'].median(),
        rendimientos_esperados_mes['Israel_monthly_exp'].median(),
        rendimientos_esperados_mes['Russia_monthly_exp'].median()
    ],
    'Desv.Std ': [
        rendimientos_esperados_mes['USA_monthly_exp'].std(),
        rendimientos_esperados_mes['Israel_monthly_exp'].std(),
        rendimientos_esperados_mes['Russia_monthly_exp'].std()
    ],
    'Mínimo ': [
        rendimientos_esperados_mes['USA_monthly_exp'].min(),
        rendimientos_esperados_mes['Israel_monthly_exp'].min(),
        rendimientos_esperados_mes['Russia_monthly_exp'].min()
    ],
    'Máximo ': [
        rendimientos_esperados_mes['USA_monthly_exp'].max(),
        rendimientos_esperados_mes['Israel_monthly_exp'].max(),
        rendimientos_esperados_mes['Russia_monthly_exp'].max()
    ],
    'Asimetría': [
        rendimientos_esperados_mes['USA_monthly_exp'].skew(),
        rendimientos_esperados_mes['Israel_monthly_exp'].skew(),
        rendimientos_esperados_mes['Russia_monthly_exp'].skew()
    ],
    'Curtosis': [
        rendimientos_esperados_mes['USA_monthly_exp'].kurtosis(),
        rendimientos_esperados_mes['Israel_monthly_exp'].kurtosis(),
        rendimientos_esperados_mes['Russia_monthly_exp'].kurtosis()
    ]
})

print(stats_rentabilidad.to_string(index=False))
````

Al abordar valores atípicos, es relevante cuestionarse sobre la naturaleza de estos puntos extremos.
* ¿Son errores de medición o representan información válida pero excepcional?
* ¿Qué criterios o técnicas se utilizaron para identificar los valores atípicos?

````python
# ---INGRESE SU CÓDIGO---
# Identificar las columnas a analizar (ignorando la fecha)
columnas_financieras = [col for col in df_finance.columns if col != 'Date']

# Iterar sobre cada país y variable
for col in columnas_financieras:
    # Aplicar la ecuación IQR para límites estadísticos
    Q1 = df_finance[col].quantile(0.25)
    Q3 = df_finance[col].quantile(0.75)
    IQR = Q3 - Q1

    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR

    # Extraer los datos atípicos
    atipicos = df_finance[(df_finance[col] < limite_inferior) | (df_finance[col] > limite_superior)]

    # 2. Generar el Boxplot (Visual)
    plt.figure(figsize=(10, 4))
    sns.boxplot(x=df_finance[col], color='lightsteelblue')

    # Añadir líneas de los límites calculados
    plt.axvline(limite_inferior, color='red', linestyle='--', label=f'Límite Inf: {limite_inferior:.4f}')
    plt.axvline(limite_superior, color='red', linestyle='--', label=f'Límite Sup: {limite_superior:.4f}')

    plt.title(f'Distribución y Valores Atípicos: {col}', fontsize=12)
    plt.xlabel('Valor')
    plt.legend()
    plt.tight_layout()
    plt.show()

    # 3. Imprimir el resumen estadístico
    print(f"--- Análisis para: {col} ---")
    print(f"Q1: {Q1:.4f} | Q3: {Q3:.4f} | IQR: {IQR:.4f}")
    print(f"Total de valores atípicos encontrados: {len(atipicos)}")

    if not atipicos.empty:
        print("Detalle de atípicos:")
        if 'Date' in df_finance.columns:
            # Mostrar la fecha para identificar el evento histórico
            print(atipicos[['Date', col]].to_string(index=False))
        else:
            print(atipicos[[col]].to_string(index=False))
    print("="*50 + "\n")
````

Los mercados financieros globales operan bajo sistemas de monitoreo sumamente precisos y rigurosos, en especifico estos datos utilizaron yfinance y otros sitemas muy precisos, por lo que es estadísticamente improbable que estos puntos extremos sean errores de captura o de medición. En cambio, son reflejos exactos de *shocks* del mercado, lo cual es bastante importante para nuestro futuro analisis. Por ejemplo, la caída extrema en el rendimiento de Israel (-0.1045) o los picos de volatilidad en Estados Unidos (0.1454) y Rusia (0.1454) pueden estar documentando momentos de alta tensión, crisis institucionales o estallidos de conflictos geopolíticos. Son anomalías numéricas, pero realidades históricas.

Para identificar los valores atípicos se utilizo **Rango Intercuartílico (IQR)** o método de Tukey. Esta técnica es ideal porque es robusta ante distribuciones asimétricas (comunes en finanzas). Se establecieron límites matemáticos calculando la diferencia entre el tercer y primer cuartil ($IQR = Q_3 - Q_1$), y marcando como atípico cualquier valor que se desviara 1.5 veces el IQR por debajo del primer cuartil o por encima del tercero. Ademas se graficaron los boxplot.

En general se aplico un metodo estadistico y un metodo visual, a continuacion un pequeño resumen del metodo:

* **Estadísticamente:** Se calcularon los cuartiles y las fronteras numéricas exactas, lo que permitió contabilizar e identificar con precisión el valor exacto de la anomalía (ej. 2 atípicos en la volatilidad de EE. UU., 0 en su rendimiento).
* **Visualmente:** Se emplearon diagramas de caja (Boxplots) para observar la dispersión general de los datos y constatar gráficamente qué tan alejados estaban estos puntos extremos del comportamiento habitual (la caja) de cada país.


Finalmente decidimos **conservar e integrar (no eliminar ni suavizar)** estos valores atípicos en el análisis.

Porque eliminar estos datos significaría borrar la evidencia empírica de los eventos políticos y macroeconómicos que estamos intentando analizar.
````python
# Se definen los países y sus columnas correspondientes
paises = {
    'Estados_Unidos': {'vol': 'Estados Unidos 🇺🇸_volatilidad', 'sent': 'Estados Unidos 🇺🇸'},
    'Israel': {'vol': 'Israel 🇮🇱_volatilidad', 'sent': 'Israel 🇮🇱'},
    'Rusia': {'vol': 'Rusia 🇷🇺_volatilidad', 'sent': 'Rusia 🇷🇺'}
}

# Diccionarios para almacenar los DataFrames en la memoria de Python
dfs_promedios = {}
dfs_modelos = {}

for pais_nombre, columnas in paises.items():
    col_vol = columnas['vol']
    col_sent = columnas['sent']

    # =========================================================
    # 1. DATASET DE PROMEDIOS
    # =========================================================
    df_promedio = df_finance_and_sent.groupby(col_sent)[col_vol].mean().reset_index()
    df_promedio.columns = ['Categoria', 'Volatilidad promedio']

    # Multiplicamos por 100 y añadimos el símbolo '%' para formato Excel
    df_promedio['Volatilidad promedio'] = (df_promedio['Volatilidad promedio'] * 100).round(2)

    # Guardamos en el diccionario
    dfs_promedios[pais_nombre] = df_promedio

    # Exportar a CSV
    nombre_archivo_promedio = f"{pais_nombre}_promedios.csv"
    df_promedio.to_csv(nombre_archivo_promedio, index=False, encoding='utf-8-sig')

    # =========================================================
    # 2. DATASET DEL MODELO (LAG1 + Dummies sin NEU)
    # =========================================================
    df_temp = df_finance_and_sent[[col_vol, col_sent]].copy()

    # Crear el LAG1
    df_temp['LAG1'] = df_temp[col_vol].shift(1)

    # Crear variables dummy
    dummies = pd.get_dummies(df_temp[col_sent], dtype=int)
    dummies.columns = [f"{col}" for col in dummies.columns]

    # Se elimina NEU
    if 'NEU' in dummies.columns:
        dummies.drop(columns=['NEU'], inplace=True)

    # Unir todo
    df_modelo = pd.concat([df_temp['LAG1'], dummies, df_temp[col_vol]], axis=1)
    df_modelo.rename(columns={col_vol: 'Volatilidad (Y)'}, inplace=True)

    # Eliminar el primer mes porque el LAG1 queda vacío
    df_modelo.dropna(inplace=True)

    # Quitar columnas dummy que no tengan datos (por ejemplo, si no hubo POS)
    for col in dummies.columns:
        if col in df_modelo.columns and df_modelo[col].sum() == 0:
            df_modelo.drop(columns=[col], inplace=True)

    # Guardamos en el diccionario
    dfs_modelos[pais_nombre] = df_modelo

    # Exportar a CSV
    nombre_archivo_modelo = f"{pais_nombre}_modelo.csv"
    df_modelo.to_csv(nombre_archivo_modelo, index=False, encoding='utf-8-sig')

    # Confirmación en pantalla para saber que funcionó correctamente
    print(f"=============================")
    print(f"--- {pais_nombre} ---")
    print(f"=============================")
    print(f"\n--- Df Modelo ---")
    print(df_modelo.head(5))
    print(f"\n--- Df promedio ---")
    print(df_promedio.head(5))
    print("\n")
````

# **1.1.2 Correlaciones**


Dado que el análisis de distribución confirmó que las variables LAG1 y Volatilidad no siguen una distribución normal en ninguno de los tres países, se utilizó el coeficiente de correlación de Spearman en lugar de Pearson. A diferencia de Pearson, Spearman no asume normalidad ya que trabaja con los rangos de los datos en lugar de los valores directamente, lo que lo hace robusto ante los altos valores identificados previamente y asociados a eventos políticos de alto impacto como el conflicto de Gaza y las sanciones sobre Rusia.

```
Datos normales = Pearson
Datos no normales o con valores extremos = Spearman
```








````python
# CORRELACIÓN DE SPEARMAN
# LAG1 vs Volatilidad (Y)

from scipy.stats import spearmanr

for pais, df in dfs_modelos.items():

    lag1 = df['LAG1'].dropna()
    vol = df['Volatilidad (Y)'].dropna()

    r,p = spearmanr(lag1, vol)

    print(f'{pais}')
    print(f'Spearman r = {r:.4f} ')
    print(f"  p-value    = {p:.4f}")

# Interpretación de la fuerza

fuerza = 'Muy fuerte' if abs(r)> 0.8 else \
         'Fuerte'     if abs(r)> 0.6 else \
         'Moderada'   if abs(r)> 0.4 else \
         'Débil'      if abs(r)> 0.2 else 'Muy débil'

direccion = 'Positiva' if r>0 else 'Negativa'

print(f'Fuerza = {fuerza}')
print(f'Dirección = {direccion}')


````

````python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# . Diccionario para guardar los modelos entrenados
modelos_entrenados = {}

# 3. Iterar sobre cada país en el diccionario
for pais, df in dfs_modelos.items():
    print(f"--- Resultados del Modelo para {pais} ---")

    # Separar variables independientes (X) de la dependiente (Y)
    X = df.drop(columns=['Volatilidad (Y)'])
    y = df['Volatilidad (Y)']

    # Crear y entrenar el modelo
    modelo = LinearRegression()
    modelo.fit(X, y)

    # Guardar el modelo en caso de que quieras hacer predicciones después
    modelos_entrenados[pais] = modelo

    # Calcular métricas y parámetros
    y_pred = modelo.predict(X)
    r_cuadrado = r2_score(y, y_pred)

    # Imprimir los coeficientes (pesos de cada variable) y el intercepto
    print(f"Intercepto (Beta 0): {modelo.intercept_:.6f}")

    coeficientes = dict(zip(X.columns, modelo.coef_))
    for variable, coef in coeficientes.items():
        print(f"Coeficiente para {variable}: {coef:.6f}")

    print(f"R-cuadrado: {r_cuadrado:.4f}\n")
````

````python

for pais, df in dfs_modelos.items():
    y = df['Volatilidad (Y)']
    X_base = df.drop(columns=['Volatilidad (Y)'])

    # Se crean aproximaciones independientes

    #  Polinómica (Agregamos solo el cuadrado)
    df_poly = X_base.copy()
    df_poly['LAG1_Cuadrado'] = df_poly['LAG1'] ** 2

    # Interacción (Agregamos solo la multiplicación)
    df_inter = X_base.copy()
    df_inter['Interaccion_LAG1_NEG'] = df_inter['LAG1'] * df_inter['NEG']

    # Logarítmica (Reemplazamos LAG1 por su logaritmo natural)
    df_log = X_base.copy()
    df_log['Log_LAG1'] = np.log(df_log['LAG1'])
    df_log = df_log.drop(columns=['LAG1'])

    # Diccionario para iterar y competir
    aproximaciones = {
        'Lineal Normal (Original)': X_base,
        'Polinómica (Curva)': df_poly,
        'Interacción (LAG1 x NEG)': df_inter,
        'Logarítmica': df_log
    }

    mejor_nombre = ""
    mejor_r2 = -float('inf')
    mejor_modelo = None
    mejor_X = None

    print(f"\n{'='*50}")
    print(f"📊 COMPETENCIA DE MODELOS: {pais.upper()}")
    print(f"{'='*50}")

    # Probando cada modelo
    for nombre, X in aproximaciones.items():
        modelo = LinearRegression().fit(X, y)
        r2 = r2_score(y, modelo.predict(X))
        print(f"  Aproximación {nombre} -> R2: {r2:.4f}")

        # Guardamos el mejor
        if r2 > mejor_r2:
            mejor_r2 = r2
            mejor_nombre = nombre
            mejor_modelo = modelo
            mejor_X = X

    # Imprimiendo el ganador y sus coeficientes limpios
    print(f"\n🏆 MEJOR APROXIMACIÓN: {mejor_nombre}")
    print(f"   R-cuadrado Ganador: {mejor_r2:.4f}")
    print("   Coeficientes del Modelo:")
    print(f"     - Intercepto (Beta 0): {mejor_modelo.intercept_:.6f}")
    for var, coef in zip(mejor_X.columns, mejor_modelo.coef_):
        print(f"     - {var}: {coef:.6f}")
````

````python
import plotly.express as px
import plotly.graph_objects as go
#Entrenar modelos ganadores y unificar datos en un solo DataFrame
datos_completos = []

# EE. UU.
df_us = dfs_modelos['Estados_Unidos'].copy()
X_us = df_us[['LAG1', 'NEG', 'POS']].copy()
X_us['Interaccion'] = X_us['LAG1'] * X_us['NEG']
modelo_us = LinearRegression().fit(X_us, df_us['Volatilidad (Y)'])
df_us['Prediccion'] = modelo_us.predict(X_us)
df_us['Pais'] = 'Estados Unidos'
df_us['Dia'] = df_us.index + 1
datos_completos.append(df_us)

# Israel
df_is = dfs_modelos['Israel'].copy()
X_is = df_is[['LAG1', 'NEG']].copy()
X_is['LAG1_Cuadrado'] = X_is['LAG1']**2
modelo_is = LinearRegression().fit(X_is, df_is['Volatilidad (Y)'])
df_is['Prediccion'] = modelo_is.predict(X_is)
df_is['Pais'] = 'Israel'
df_is['Dia'] = df_is.index + 1
# Para que las columnas coincidan en el DataFrame final, añadimos POS e Interaccion vacías
df_is['POS'] = 0
datos_completos.append(df_is)

# Rusia
df_ru = dfs_modelos['Rusia'].copy()
X_ru = df_ru[['LAG1', 'NEG']].copy()
X_ru['LAG1_Cuadrado'] = X_ru['LAG1']**2
modelo_ru = LinearRegression().fit(X_ru, df_ru['Volatilidad (Y)'])
df_ru['Prediccion'] = modelo_ru.predict(X_ru)
df_ru['Pais'] = 'Rusia'
df_ru['Dia'] = df_ru.index + 1
df_ru['POS'] = 0
datos_completos.append(df_ru)

# Unimos todo en una sola tabla
df_final = pd.concat(datos_completos, ignore_index=True)

# VISUALIZACIÓN INTERACTIVA 1: SCATTER PLOT GENERAL

fig1 = px.scatter(
    df_final,
    x='Volatilidad (Y)',
    y='Prediccion',
    color='Pais',
    hover_data=['Dia', 'LAG1', 'NEG'], # Datos extra al pasar el ratón
    title='Interactivo: Volatilidad Real vs. Predicha por País',
    labels={'Volatilidad (Y)': 'Volatilidad Real (Observada)', 'Prediccion': 'Volatilidad Predicha por el Modelo'},
    template='plotly_white',
    size_max=15
)

# Añadimos la línea ideal de predicción perfecta
lim_min = min(df_final['Volatilidad (Y)'].min(), df_final['Prediccion'].min())
lim_max = max(df_final['Volatilidad (Y)'].max(), df_final['Prediccion'].max())
fig1.add_shape(
    type="line", line=dict(dash='dash', color="gray"),
    x0=lim_min, y0=lim_min, x1=lim_max, y1=lim_max
)
fig1.update_traces(marker=dict(size=12, line=dict(width=1, color='DarkSlateGrey')))
fig1.show()


# VISUALIZACIÓN INTERACTIVA 2: LÍNEA DE TIEMPO (ESTADOS UNIDOS)

df_usa_only = df_final[df_final['Pais'] == 'Estados Unidos']

fig2 = go.Figure()
# Línea de Realidad
fig2.add_trace(go.Scatter(x=df_usa_only['Dia'], y=df_usa_only['Volatilidad (Y)'],
                    mode='lines+markers', name='Volatilidad Real',
                    line=dict(color='blue', width=3), marker=dict(size=10)))
# Línea de Predicción
fig2.add_trace(go.Scatter(x=df_usa_only['Dia'], y=df_usa_only['Prediccion'],
                    mode='lines+markers', name='Predicción del Modelo',
                    line=dict(color='orange', width=3, dash='dash'), marker=dict(size=10)))

fig2.update_layout(
    title='Seguimiento Temporal: Realidad vs. Predicción en Estados Unidos (Mejor Modelo)',
    xaxis_title='Día de Observación (1 al 10)',
    yaxis_title='Nivel de Volatilidad',
    template='plotly_white',
    hovermode="x unified" # Muestra ambos valores al pasar el ratón por un día
)
fig2.show()
````

### **2.4. Interpretación de Resultados**

1. ¿Existen patrones o tendencias significativas en los datos?

El análisis descriptivo revela que la relación entre política y mercado es contextual y regional . En EE.UU. , existe un patrón de "sensibilidad acumulativa": la volatilidad no reacciona solo a la noticia negativa per se , sino que esto actúa como un catalizador que potencia la inestabilidad previa. En contraste, en Israel , se identificó una elevación directa y significativa entre el sentimiento de la noticia y el índice TA-35, confirmando que los eventos políticos son determinantes primarios de su riesgo de mercado. Mientras que el mercado estadounidense tiende a la persistencia de volatilidad, los mercados de Israel y Rusia muestran una dinámica de "shocks externos" donde los eventos políticos sobreescriben los ciclos históricos, aunque en el caso israelí se observa una tendencia a la estabilización tras el impacto inicial.

2. ¿Qué significan los resultados en términos prácticos?

Desde una perspectiva de explicabilidad, los resultados demuestran que el riesgo de mercado no es un fenómeno puramente financiero , sino que está profundamente anclado a la estabilidad institucional. En Israel, la alta sensibilidad a las crisis políticas traduce la incertidumbre gubernamental en una prima de riesgo inmediato y mayor probabilidad de pérdidas. En el caso norteamericano, el mercado se comporta de forma más lógica y predecible ante el contexto geopolítico: la calma política facilita la resiliencia, mientras que la agitación política vuelve al mercado vulnerable a cualquier flujo de información. En resumen, la "geografía del riesgo" dicta que en economías con conflictos activos, las variables políticas son fuerzas masivas que invalidan los patrones de comportamiento técnico tradicional.
3. ¿Existen patrones o tendencias inesperadas en los datos que requerirán una consideración especial o una investigación adicional?

Se halló una anomalía matemática en el modelo de EE.UU. el coeficiente negativo para noticias negativas sugiere que, ante ciertos eventos políticos, el mercado podría experimentar un "periodo de parálisis" o contracción temporal de la volatilidad antes de reaccionar con pánico. Este hallazgo desafió la intuición y requiere investigar si se trata de un comportamiento de cautela institucional. Asimismo, la rápida capacidad de corrección en Israel tras eventos políticos es un patrón atípico que sugiere una resiliencia estructural digna de un estudio más profundo. Finalmente, la falta de significancia estadística en algunos cruces de datos entre Rusia y EE.UU. apunta a la necesidad de ampliar la muestra de eventos políticos para diferenciar entre ruido aleatorio y verdaderas tendencias de riesgo sistémico.


4. ¿Cuáles fueron los hallazgos principales que contribuirán al proceso de toma de decisiones?


La investigación proporciona evidencia cuantitativa para tres pilares de decisión:
* **Diferenciación de Dinámicas:** La gestión del riesgo debe ser específica para cada país; no se puede aplicar la misma lógica de "reacción a la noticia" en EE.UU. que en Israel.
* **Umbrales de Alerta:** En mercados desarrollados, la volatilidad previa es el mejor predictor de cómo afectará una noticia política futura, permitiendo establecer protocolos de cobertura basados ​​en el histórico inmediato.
* **Límites de los Modelos Tradicionales:** Se demuestra que, en escenarios de conflicto o crisis de gobierno (como en Rusia e Israel), los modelos basados ​​únicamente en precios pasados ​​son insuficientes. La toma de decisiones debe integrar obligatoriamente el análisis de sentimiento político como una variable explicativa fundamental para entender la exposición real al riesgo.
